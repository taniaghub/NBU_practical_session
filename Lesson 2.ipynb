# 2. Open and read web-page

2.1. Import modules

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
```

2.2. Set URL

```python
url = "https://www.ikea.com/ua/uk/cat/tovari-products/"
print (url)
```

2.3. Request page information

```python
#some pages have protection from robots, but Python provides useful tool - it can open pages as web-browser, imitating human behaviour
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
f = Request(url, headers=headers)
myfile = urlopen(f).read()
print (myfile)
```

2.4. Apply BeautifulSoup

```python
TopPageSoup = BeautifulSoup(myfile)
print (TopPageSoup)
```

# 3. Scrape information

3.1. Find data blocks with link information

```python
product = TopPageSoup.findAll('a', attrs={'class':'vn-link vn-nav__link'})
print (product)
```

3.2. Select link name from string:

    - go to https://regex101.com and select Python

    - copy any string from above and insert it into "Test string" window

    - find regular expresion that fits

```python
for i in range(0,len(product)): #iterate all data blocks one by one
    name = re.search("href=\"(.*?)\">", str(product[i]).replace("\n","").replace(";","")).groups(0)[0]
    print (name)
```

3.3. Repeat everything on the next level

```python
#request and open page
p = Request(name, headers=headers)
myfile = urlopen(p, timeout=30).read() 
#apply BeautifulSoup
TopPageSoup = BeautifulSoup(myfile)
#find data blocks with product information
goods = TopPageSoup.findAll('div', attrs={'class':'range-revamp-product-compact__bottom-wrapper'})
print (goods)
```

3.4. Scrape links for every unique product

```python
for j in range(0,len(goods)):
    good_link = re.search("a href=\"(.*?)\">", str(goods[j]).replace("\n","").replace(";","")).groups(0)[0]
    print (good_link)
```

3.5. Now we can read page with product information

```python
g = Request(good_link, headers=headers)
myfile = urlopen(g, timeout=30).read()
TopPageSoup = BeautifulSoup(myfile)
print (TopPageSoup)
```

3.6. Convert unstructured data to structured 

```python
#find good name
good = TopPageSoup.findAll('h1', attrs={'class':'range-revamp-header-section'})
print (good)
good_name=re.search("big notranslate\">(.*?)\<", str(good).replace("\n","").replace(";","")).groups(0)[0] #be careful with groups
print ("\n",good_name) #"\n" indicates new line
```

```python
#find good price
price_full = TopPageSoup.findAll('span', attrs={'class':'range-revamp-price__integer'})
print (price_full)
price = re.search("integer\">(.*?)\<", str(price_full).replace("\n","").replace(" ","").replace(";","")).groups(0)[0]
print ("\n",price)            
```

```python
#find good category
cat = TopPageSoup.findAll('li', attrs={'class':'bc-breadcrumb__list-item'})
print (cat)
category_full = re.search("span>(.*?)<\/span", str(cat[2]).replace("\n","").replace(";","")).groups(0)[0]
print ("\n",category_full)
category = re.search("span>(.*?)<\/span", str(cat[3]).replace("\n","").replace(";","")).groups(0)[0]
print ("\n",category)         
```

```python
#find additional useful information
#product_id is useful for international comparisons
id_n=TopPageSoup.findAll('script', attrs={'data-type':'utag-data'})
print (id_n)
product_id = re.search("product_ids\"\:\[\"(.*?)\"]", str(id_n).replace("\n","").replace(";","")).groups(0)[0]   
print ("\n",product_id)
#we also scrape data on id of category
category_int = re.search("category\"\:\"(.*?)\"", str(id_n).replace("\n","").replace(";","")).groups(0)[0]
print ("\n",category_int)
#you can add everything you find useful on the page, for example this site have information about VAT
tax= re.search("tax\"\:\[\"(.*?)\"\]", str(id_n).replace("\n","").replace(";","")).groups(0)[0]
print ("\n",tax)
```

```python
#Let's check start date
dt_object = dt.fromtimestamp(0) #everything 
print (dt_object)
#and current date
dt_object = dt.fromtimestamp(time.time()) #shows your local time
print (dt_object)
```

5.4. Find starting point of the day

```python
#round down the current day number
day_start=int(int(time.time())/60/60/24)*60*60*24
print (day_start)
```

```python
#check what we get
dt_object = dt.fromtimestamp(day_start)
print (dt_object)
```

5.5. Add one day to start point

```python
#we can add more than 60*60*24 seconds if we want our program to start later
next_day_start=int(int(time.time())/60/60/24)*60*60*24+60*60*24
print (next_day_start)
```

```python
#check what we get
dt_object = dt.fromtimestamp(next_day_start)
print (dt_object)
```

5.6. Operator while

```python
i = 0 
while i < 3: #repeats actions infinite times until error occurs or condition becomes False
    dt_object = dt.fromtimestamp(time.time())
    print (i, dt_object)
    pause.until(int(int(time.time())/60)*60+60)
    i = i+1
```
